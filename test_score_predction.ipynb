```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

#creating data set
student_data = {
    'Study Hours' : [2, 3, 4, 5, 8, 9, 20, 30, 35,40 ],
    'Test Score' : [60, 5, 34, 44, 84, 76, 56, 57, 98, 100]
}

#creating a pandas

df = pd.DataFrame(student_data)

#preparing futures and target
X = df[['Study Hours']] #futures
Y = df['Test Score'] #Target

# Splitting Data into Training and Testing Sets

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)

# Training the Linear Regression Model
model = LinearRegression()
model.fit(X_train, Y_train)

# Making Predictions
Y_pred = model.predict(X_test)

#Evaluate
mse = mean_squared_error(Y_test, Y_pred)
print(f"MSE: {mse:.2f}")
print(f"Slope: {model.coef_[0]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")

# Visualize
plt.scatter(X_train, Y_train, color='orange', label='Data Points')
plt.plot(X, model.predict(X), color='blue', label='Regression Line')
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.title('Study Hours vs. Test Score')
plt.legend()
plt.show()

# Predict new value
new_hours = pd.DataFrame([[7.5]], columns=['Study Hours'])
predicted_score = model.predict(new_hours)
print(f"Predicted score for 7.5 study hours: {predicted_score[0]:.2f}")
```
